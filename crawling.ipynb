```python
for i in range(len(df20)):
    elem = driver.find_element(By.XPATH, '//*[@id="top_search"]')
    elem.send_keys(df20.loc[i, 'song'] + ' ' + df20.loc[i, 'singer'] )
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="gnb"]/fieldset/button[2]/span').click()
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="divCollection"]/ul/li[3]/a/span').click()
    time.sleep(1)
    driver.find_element(By.XPATH, '//*[@id="frm_defaultList"]/div/table/tbody/tr[1]/td[3]/div/div/a[1]/span').click()
    time.sleep(0.5)
    
    html = driver.page_source # 드라이버 현재 페이지의 html 정보 가져오기 
    soup = BeautifulSoup(html, 'lxml')
    time.sleep(0.5)
    comment = soup.select('#revCnt')[0].text
    date = soup.select('#downloadfrm > div > div > div.entry > div.meta > dl > dd:nth-child(4)')[0].text

    if cnt % 50 == 0:
        pd.DataFrame(result_list).to_csv(f'data{cnt}.csv')
    cnt += 1
    song_comment = []

    driver.find_element(By.XPATH, '//*[@id="revCnt"]').click()
    j = 1
    while True:
        try:
            html = driver.page_source # 드라이버 현재 페이지의 html 정보 가져오기 
            soup = BeautifulSoup(html, 'lxml')
            time.sleep(0.5)
            song_comment.append(soup.select('#d_cmtpgn_cmt_list_wrapper > ul > li.first_child > div > div.wrap_cntt.d_cmtpgn_cmt_cont_wrapper > div.cntt > div')[0].text.split('\t')[-2])
            for k in range(2,11):
                song_comment_elem = soup.select(f'#d_cmtpgn_cmt_list_wrapper > ul > li:nth-child({k}) > div > div.wrap_cntt.d_cmtpgn_cmt_cont_wrapper > div.cntt > div')
                if song_comment_elem:
                    song_comment.append(song_comment_elem[0].text.split('\t')[-2])
                else:
                    continue
            driver.find_element(By.XPATH, f'//*[@id="d_cmtpgn_paginate_wrapper"]/span/a[{j}]').click()
            time.sleep(0.5)
            driver.refresh()
            j += 1
        except :
            next_button = driver.find_element(By.CSS_SELECTOR, "#d_cmtpgn_paginate_wrapper > a.btn_next.d_cmtpgn_navigator")
            if "disabled" in next_button.get_attribute("class"):
                # "다음" 버튼이 비활성화되면 종료
                break
            else:
                next_button.click()
                j = 1
                time.sleep(1)
                continue
    fan_cnt = 0
    try:
        for art in range(1, len((driver.find_element(By.XPATH, '//*[@id="downloadfrm"]/div/div/div[2]/div[1]/div[2]').text.split(',')))+1):
            driver.find_element(By.XPATH, f'//*[@id="downloadfrm"]/div/div/div[2]/div[1]/div[2]/a[{art}]').click()
            html = driver.page_source # 드라이버 현재 페이지의 html 정보 가져오기 
            soup = BeautifulSoup(html, 'lxml')
            fan_cnt += int(soup.select('#conts > div.wrap_dtl_atist > div > div.wrap_intst > div.fan_area > span > span')[0].text[:-1].replace(',',''))
            driver.back()
    except :
        pass


        driver.back()
        
    result_list.append(
        {
            'singer' : df20.loc[i, 'singer'],
            'song' : df20.loc[i, 'song'],
           'comment_cnt' : comment,
           'song_comment' : song_comment,
            'fan_cnt' : fan_cnt,
            'release' : date
        }
    )

    time.sleep(0.5)
    
driver.close()
```
